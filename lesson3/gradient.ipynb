```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

####  图像边缘提取
#### 1.Sobel算法

```python
kernel=np.ones((3,3),np.uint8)
img=cv2.imread(r'E:\pythonProject\opencvProject\resources\girlfriend1.jpg',0)
sobelx=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
sobely=cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
sobel=cv2.Sobel(img,cv2.CV_64F,1,1,ksize=5)
laplacian=cv2.Laplacian(img,ddepth=cv2.CV_64F)
plt.subplot(221),plt.imshow(img,'gray'),plt.title('sobelx')
plt.xticks([]),plt.yticks([])
plt.subplot(222),plt.imshow(sobely,'gray'),plt.title('sobely')
plt.xticks([]),plt.yticks([])
plt.subplot(223),plt.imshow(sobel,'gray'),plt.title('sobel')
plt.xticks([]),plt.yticks([])
plt.subplot(224),plt.imshow(laplacian,'gray'),plt.title('laplacian')
plt.xticks([]),plt.yticks([])

plt.show()
```

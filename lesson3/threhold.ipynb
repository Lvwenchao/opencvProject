```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread(r'E:\pythonProject\opencvProject\resources\bandw.jpg')
```

#### 1.threhold() ----- 简单阈值


```python
ret,thresh1=cv2.threshold(img,125,255,cv2.THRESH_BINARY)
ret,thresh2=cv2.threshold(img,125,255,cv2.THRESH_BINARY_INV)
ret,thresh3=cv2.threshold(img,125,255,cv2.THRESH_TRUNC)
ret,thresh4=cv2.threshold(img,125,255,cv2.THRESH_TOZERO)
ret,thresh5=cv2.threshold(img,125,255,cv2.THRESH_TOZERO_INV)
imgs=[img,thresh1,thresh2,thresh3,thresh4,thresh5]
titles=['original','binary','binary_inv','trunk','tozero','tozero_inv']
for i in range(len(titles)):
    plt.subplot(2,3,i+1),plt.imshow(imgs[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
# cv2.imshow('image', thresh1)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

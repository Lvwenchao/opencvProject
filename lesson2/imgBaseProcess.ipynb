```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r'E:\pythonProject\opencvProject\resources\head(200_200).jpg', 1)
```

#### 1.修改像素

```python
print(img.shape)
img[:100,:100]=[21,212,21]# 20*20*3 的数据
```

```python
# 移动物体
head=img[20:160,40:180]
img[60:200,50:190]=head
```

#### 2.拆分合并通道

```python
b,g,r=cv2.split(img)
print(b.shape)
img = cv2.merge([b,g,r])
```

#### 3.为图像扩边


```python
replicate=cv2.copyMakeBorder(img,10,10,10,10,borderType=cv2.BORDER_REPLICATE)
reflect=cv2.copyMakeBorder(img,10,10,10,10,borderType=cv2.BORDER_REFLECT)
reflect101=cv2.copyMakeBorder(img,10,10,10,10,borderType=cv2.BORDER_REFLECT101)
constant=cv2.copyMakeBorder(img,10,10,10,10,borderType=cv2.BORDER_CONSTANT,value=[255,0,0])
plt.subplot(231),plt.imshow(replicate,'gray'),plt.title('REPLICATE')
plt.subplot(232),plt.imshow(reflect,'gray'),plt.title('REFLECT')
plt.subplot(233),plt.imshow(reflect101,'gray'),plt.title('REFLECT_101')
plt.subplot(234),plt.imshow(constant,'gray'),plt.title('CONSTANT')
```

```python
cv2.namedWindow('godfather', cv2.WINDOW_NORMAL)
cv2.imshow('godfather', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### 4.尺寸修改


```python
img2=cv2.imread(r'E:\pythonProject\opencvProject\resources\2(80_80).jpg')
img=cv2.resize(img,(img2.shape[0],img2.shape[1]),interpolation=cv2.INTER_CUBIC)
cv2.imshow('godfather', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
